# Облигации краткая инфа.
В целом облигации - это ты даешь в долг.
Цена обычна в процентах от номинала. Номинал чаще всего 1000. Если номинал меньше, значит уже частично начали погашать.

Может быть и 10к и 100к, но там надо прям хорошо серчить про компании. Так что лучше на 1к сидеть.
Не важно засколько ты купила облигацию, в конце срока ее погасят по фул цене. (иногда это чисто один платеж в конце, а иногда как на фото выше.)

Листинг 1 или 2 -берем, но точно не покупать 3 (есть исключения для спикуляции). Вообще листинг- это уровень надежности бумаги по мнению биржи 1- самый надежный.

НКД - накопленный купонный доход. Ты его платишь чтобы при ближайшей выплате купона получить его полностью. Соответственно когда продаешь облигацию тебе тоже выплатят  этот НКД.

По купонам лучше брать только фиксированные, чтобы не зависили от ставки.

Надо смотреть еще на наличие (Этого нет в общем доступе. Если что кидай isin облигации я быстро гляну):
- дефолта (Эмитент не в состоянии выполнить свои обязательства по выплате процентов или погашению долга в срок)
- амортизации (номинальная стоимость возвращается не целиком в дату погашения, а частями вместе с выплатой купонов /как на фото/)
- оферты (условия без которых нельзя продать)
